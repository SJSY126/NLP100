```python
def parse_line(line):
    (surface, temp) = line.split("\t")
    temp = temp.split(",")
    return{
        'surface':surface,
        'base':temp[6],
        'pos':temp[0],
        'pos1':temp[1]
    }
    

def load_mecab_file(filename):
    """
    filname: mecabファイルパス
    """
    sentence = []
    sentences = []
    with open(filename) as f:
        for line in f:
            line = line.replace("\n", "")
            if line == "EOS":
                if len(sentence) > 0:
                    sentences.append(sentence)
                sentence = []
            else:
                sentence.append(parse_line(line))
    return sentences
```

```python
sentences = load_mecab_file('./neko.txt.mecab-ipadic')

word_count = {}

for sentence in sentences:
    for word in sentence:
        temp = '{}\t{}\t{}'.format(
            word['base'],
            word['pos'],
            word['pos1']
        )
        if temp in word_count:
            word_count[temp] += 1
        else:
            word_count[temp] = 1
            
word_count_sorted = sorted(word_count.items(), key=lambda x:x[1], reverse=1)
word_count_sorted
```
